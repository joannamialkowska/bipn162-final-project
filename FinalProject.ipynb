```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
adata = sc.read_h5ad('c2c0a633-136f-4c97-ab51-062ce97ad802.h5ad')
gene_df = adata.to_df()
gene_df.index = gene_df.index.str.replace('-.*$', '', regex=True)
gene_df = gene_df.iloc[:, np.flatnonzero(gene_df.sum(axis = 0) != 0)]

print(f"There are {gene_df.shape[0]} cells and {gene_df.shape[1]} genes in the count matrix:")

# Show the first 5 rows of the count matrix
gene_df.head()
```

```python
ephys_df = pd.read_csv('m1_patchseq_ephys_features.csv').set_index('cell id')
morph_df = pd.read_csv('m1_patchseq_morph_features.csv').set_index('cell id')
merged_data = pd.merge(ephys_df, morph_df, left_index=True, right_index=True)
final_data = pd.merge(merged_data, gene_df, left_index = True, right_index = True)
final_data.head()
```
